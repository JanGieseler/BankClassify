# Label data

### load data and create classifier

```python
bc = BankClassify(Path('../data'))
```

```python
data_unlabeled = bc.data_unlabeled
print(f"{len(data_unlabeled)} unlabeled entries")
predictions = bc._make_predictions(data_unlabeled)
```

```python
# predictions.columns = ['class_guess', 'class_prob']
# predictions
```

### make guess

##### order by class probability

```python
data_unlabeled_with_guess = pd.concat([data_unlabeled, predictions], ignore_index=False, axis=1, copy=False, join='inner')
data_unlabeled_with_guess.sort_values('class_prob', inplace=True) # sort such that the ones with the highest uncertainty come first
data_unlabeled_with_guess
```

##### order by value

```python
data_unlabeled_with_guess = pd.concat([data_unlabeled, predictions], ignore_index=False, axis=1, copy=False, join='inner')
data_unlabeled_with_guess.sort_values('amount', inplace=True, ascending=False) # sort such that the ones with the highest uncertainty come first
data_unlabeled_with_guess
```
